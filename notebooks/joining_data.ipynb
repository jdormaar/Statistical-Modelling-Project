```python
def convert_to_timezone(isoFormat, defaultTz = "US/Eastern"):
  # Inconsistent date time formats need to be parsed
  date_patterns = ['%Y-%m-%dT%H:%M:%S.%fZ', '%Y-%m-%dT%H:%M:%SZ']

  for pattern in date_patterns:
    try:
      date = datetime.strptime(isoFormat, pattern)
      return date.astimezone(tz=pytz.timezone(defaultTz))
    except:
      pass

for i in range(dfcb.shape[0]):
  utc = dfcb['timestamp'].iloc[i]
  dfcb['timestamp'].iloc[i] = convert_to_timezone(utc)

dfcb.head(2)
```
